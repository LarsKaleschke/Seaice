# Snow albedo

## Theory
Scattering Optics of Snow, A. Kokhanovsky and E. P. Zege, Appl. Optics 2004 43(7)

## Data

Warren, S. G., 1984: Optical constants of ice from ultraviolet to microwave, Applied Optics, 23, 1206-1225


```python
%pylab inline

def refl_snow(wavelength,absorp,d):
    kappa=4*pi*(absorp/wavelength)
    b=3.6
    r=exp(-b*sqrt(kappa*d))
    return r
    

filename='ice_1995.dat'
a=loadtxt(filename)

absorp=a[:,2]
wavelength=a[:,0]/1e6

diameters=[0.1,0.2,0.3,0.4]
for d in diameters:
    plot(wavelength*1e6,refl_snow(wavelength,absorp,d/1000.0),label='d='+str(d)+' mm')

axis([0.4, 2.6, 0.0, 1.0])
xlabel('wavelength [$\mu$m]')
ylabel('spherical albedo')
title('Grain Size Dependend Snow Albedo')
legend(loc='upper right')
show()
```
